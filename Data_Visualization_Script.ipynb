```python
import matplotlib.pyplot as plt
import numpy as np

# DATA INPUT
methods = ['FIFO', 'Optimal VOQ', 'iSLIP VOQ']
total_times = [11, 7, 10]

# Backlog Data
fifo_backlog = [4, 4, 4, 5, 6, 6, 5, 3, 2, 1, 0, 0] 
optimal_backlog = [4, 4, 3, 3, 3, 3, 0, 0]
islip_backlog = [3, 3, 3, 4, 5, 6, 4, 2, 1, 0, 0]

# GRAPH 1: TOTAL SERVICE TIME (Bar Chart) ---
plt.figure(figsize=(8, 5))
colors = ['#e74c3c', '#2ecc71', '#3498db'] # Red, Green, Blue
bars = plt.bar(methods, total_times, color=colors, edgecolor='black', alpha=0.8)

plt.title('Performance Comparison: Total Service Time', fontsize=14, pad=15)
plt.ylabel('Total Time Slots (t)', fontsize=12)
plt.ylim(0, max(total_times) + 3)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Adding labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, yval, ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

# GRAPH 2: BACKLOG OVER TIME (Line Graph) ---
plt.figure(figsize=(10, 6))

# Plotting each line with different markers for clarity
plt.plot(fifo_backlog, label='FIFO (HoL Blocking)', color='#e74c3c', marker='o', linewidth=2)
plt.plot(optimal_backlog, label='Optimal VOQ (Theoretical)', color='#2ecc71', marker='s', linewidth=2)
plt.plot(islip_backlog, label='iSLIP VOQ (Practical)', color='#3498db', marker='^', linewidth=2)

plt.title('Switch Drain Rate: Backlog over Time', fontsize=14, pad=15)
plt.xlabel('Time Slots (t)', fontsize=12)
plt.ylabel('Number of Packets Remaining', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, which='both', linestyle=':', alpha=0.5)
plt.xticks(np.arange(0, max(total_times) + 1, 1)) # Ensure every time slot is marked

plt.tight_layout()
plt.show()
```
